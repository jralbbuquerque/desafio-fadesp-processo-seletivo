A coluna "Postal Code" possui diversas linhas com valores **nulos/vazios**. Acredito que a melhor opção seja preencer os valores com a **String** *"Missing Information"* ao invés de apagar as linhas, já que, trata-se de um Dataset com informações sobre compras de clientes.

## 3.1) Lidando com dados inconsistentes

```python
df['Postal Code'].info()
```

- O tipo da coluna **Postal Code** está em **float64**. Segundo pesquisas, o correto é mudar o tipo da coluna para **String**, já que os códigos postais não são exibidos em **float** (123456.0) e colocá-los em **interger** poderia causar problemas com algum 0 à esquerda **(009675.0 (float) == 9675 (interger))**.
- A questão é que, quando colocamos em **String**, os valores qeu terminam com **'.0'** seguem assim. Portanto, iremos modificar toda e qualquer String que termine com '.0'

```python
# Convertendo o tipo da coluna de float para String e editando as Strings que terminam em '.0'

df['Postal Code'] = df['Postal Code'].astype(str).str.replace(r'\.0$', '', regex=True)
```

## 3.2) Repondo os valores "nan" em "Missing Information"

```python
df['Postal Code'] = df['Postal Code'].replace('nan', 'Missing Information', regex=True)
```

## Engenharia de Atributos

```python
# Separando a coluna 'Order Date' em 4 colunas referentes ao ano, mês, dia e dia da semana

locale.setlocale(locale.LC_ALL, 'pt_BR.utf8')

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date'] = df['Order Date'].dt.date
df['Year'] = pd.DatetimeIndex(df['Order Date']).year
df['Month'] = pd.DatetimeIndex(df['Order Date']).month
df['Day'] = pd.DatetimeIndex(df['Order Date']).day
df['Weekday'] = pd.DatetimeIndex(df['Order Date']).day_name(locale = 'pt_BR.utf8')
```

```python
import sqlite3
from prettytable import from_db_cursor
```

```python
conn = sqlite3.connect('market_data.db')
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("DROP TABLE sales")
```

```python
cursor.execute('''CREATE TABLE sales (
                    Order_Date date,
                    Ship_Date date,
                    Month integer,
                    Year integer,
                    Customers text,
                    State text,
                    Products text,
                    Sales real,
                    Quantity integer,
                    Discount real,
                    Profit real,
                    Shipping_Cost real,
                    Ship_Mode text,
                    Order_Priority text,
                    Delivery_Time integer)''')
```

```python
schema = {
    'Order_Date' : 'DATE',
    'Ship_Date' : 'DATE',
    'Month' : 'INTEGER',
    'Year' : 'INTEGER',
    'Customers' : 'TEXT',
    'State' : 'TEXT',
    'Products' : 'TEXT',
    'Sales' : 'REAL',
    'Quantity' : 'INTEGER',
    'Discount' : 'REAL',
    'Profit' : 'REAL',
    'Shipping_Cost' : 'REAL',
    'Ship_Mode' : 'TEXT',
    'Order_Priority' : 'TEXT',
    'Delivery_Time' : 'TEXT'
}
```

```python
analise.to_sql('sales', conn, if_exists='replace', index=False, dtype=schema)
```

```python
conn.commit()
```

```python
# Método para melhor visualização das consultas em SQLite

def display_table(cursor):
    x = from_db_cursor(cursor)
    print(x)
```

```python
cursor.execute("SELECT * FROM Sales LIMIT 10")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM Sales LIMIT 10")
display_table(cursor)
```

```python
cursor.execute("SELECT Customer_Name FROM Sales LIMIT 10")
display_table(cursor)
```

```python
# Consultando os clientes que geraram lucro entre 2500 e 1000 no ano de 2012 que moram em Nova York

cursor.execute('''SELECT Customer_Name, SUM(Profit) AS Lucro 
                  FROM Sales WHERE State = 'New York' AND Year = 2012
                  GROUP BY Customer_Name
                  HAVING Lucro < 2500 AND Lucro > 1000
                  ORDER BY 2 DESC LIMIT 10;''')
display_table(cursor)
```

```python
cursor.execute('''SELECT Year, ROUND(SUM(Profit), 2) AS Lucro
                  FROM Sales
                  GROUP BY Year
                  ORDER BY 1 DESC;''')
display_table(cursor)
```
